# DATASET FORMAT

## Dataset consist of first name, last name, age, Gender, Height (in metres), Weight( in kilograms).
## To find Body Mass Index (BMI) we must divide the weight (in kilograms) by Square of Height (in metres)
## Excel formula to calculate BMI
## =[@[Weight(in kg)]]/([@[Height (in metre)]]^2)

# ABOUT THE PROBLEM:

## Body Mass Index (BMI) is a person's weight in kilogramns divided by the square of the height in meters. A high BMI can indicate high body fatness. BMI screens for weight categories that may lead to health problems, but it does not diagnose the body fatness or health of an indivisual.
